```python
#Introdution of seaborn
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
tip_df=sns.load_dataset('tips')
tip_df.head()
```

```python
sns.histplot(data=tip_df['total_bill'])
```

```python
#kiểm tra xem trong khoảng 10$ đến 30$ món nào mua nhiều
tip_df['total_bill'].value_counts().sort_values(ascending=False)
```

```python
#KDE
sns.kdeplot(data=tip_df['total_bill']);
```

```python
##displot
sns.displot(data=tip_df,x="total_bill",col='time',kde=True)
```

```python
#barplot
sns.barplot(data=tip_df,x='sex',y='tip',estimator=np.mean)
```

```python
tip_df['sex'].value_counts()
```

```python
sns.countplot(data=tip_df,x='sex')
```

```python
tip_df['smoker'].value_counts()
```

```python
sns.countplot(data=tip_df,x='smoker')
```

```python
#box plot
sns.boxplot(data=tip_df,x='day',y='total_bill',hue='sex') #hue là option thêm
```

```python
#facet grid
tip_df.head
```

```python
tip_fg=sns.FacetGrid(data=tip_df,row='smoker',col='time')
tip_fg.map(sns.scatterplot,'total_bill','tip')
```

```python
#join plot
penguins_df=sns.load_dataset('penguins')
penguins_df.head()
```

```python
sns.jointplot(data=penguins_df,x='bill_length_mm',y='flipper_length_mm',hue='species')
```

PairPlot

```python
sns.pairplot(data=penguins_df,hue='species')
```

Heatmap

```python
flights_df=sns.load_dataset('flights')
flights_df.head()
```

```python
flights=pd.pivot_table(data=flights_df,index='month',columns='year',values='passengers')
flights
```

```python
sns.heatmap(data=flights)
```

```python
diamonds=sns.load_dataset('diamonds')
diamonds
```

```python
fg,ax=plt.subplots(figsize=(6,6))
sns.despine(fg,left=True,bottom=True)
clarity_ranking=["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(data=diamonds,x='cut',y='price',hue='clarity',
                size='depth', hue_order=clarity_ranking,alpha=0.2)
```
