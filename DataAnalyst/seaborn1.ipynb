```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
cereal_df=pd.read_csv('cereal.csv')
cereal_df
```

```python
#loại bỏ các cột không liên quan
fields=['shelf','weight','cups','rating','name','mfr','type']
cereal_df_new=cereal_df.drop(fields,axis=1)
cereal_df_new.head()
```

```python
#so sánh độ tương quan
cereal_corr=cereal_df_new.corr()
cereal_corr
```

```python
#build boolen True False
ones_corr=np.ones_like(cereal_corr,dtype='bool')
ones_corr
```

```python
ones_corr.shape, cereal_corr.shape
```

```python
mask=np.triu(ones_corr)
mask
```

```python
sns.heatmap(data=cereal_corr,mask=mask)
```
