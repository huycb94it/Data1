```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities=pd.read_csv('california_cities.csv')
```

```python
cities.head()
plt.style.available
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
#tách latd: vĩ độ. longd:kinh độ
lat,long=cities['latd'],cities['longd']
population,area=cities['population_total'],cities['area_land_km2']
plt.figure(figsize=(8,6))
plt.scatter(lat,long,c=np.log10(population),cmap='viridis',s=area,alpha=0.5,linewidths=0)
plt.axis('equal')
plt.xlabel('latd')
plt.ylabel('long')
plt.colorbar(label='log$_{10}$(population)')
```
